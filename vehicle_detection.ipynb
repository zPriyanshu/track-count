```python
model = YOLO('yolo11x.pt')

line_y = 287  # y-coordinate of your counting line
count = 0
tracker_memory = {}
counted_ids = set()

#  FIX: Use stream=True and iterate results alongside video frames
results = model.track(source="xyz.mp4", persist=True, conf=0.25,
                      iou=0.5, tracker="bytetrack.yaml", stream=True)

for result in results:
    frame = result.orig_img.copy()  #  Get the actual frame from result directly

    if result.boxes.id is not None:
        boxes = result.boxes.xyxy.cpu().numpy()
        ids = result.boxes.id.cpu().numpy().astype(int)

        for box, obj_id in zip(boxes, ids):
            x1, y1, x2, y2 = map(int, box)
            cx, cy = (x1 + x2) // 2, (y1 + y2) // 2

            # Crossing logic
            if obj_id in tracker_memory:
                prev_cy = tracker_memory[obj_id]
                if prev_cy < line_y and cy >= line_y:
                    if obj_id not in counted_ids:
                        count += 1
                        counted_ids.add(obj_id)

            tracker_memory[obj_id] = cy

            # Draw bounding box and ID
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, f"ID: {obj_id}", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
            cv2.circle(frame, (cx, cy), 4, (0, 0, 255), -1)

    # Draw counting line and count — OUTSIDE the if block 
    cv2.line(frame, (0, line_y), (frame.shape[1], line_y), (0, 255, 255), 2)
    cv2.putText(frame, f"Total Vehicles: {count}", (30, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)

    cv2.imshow("YOLO Tracking & Counting", frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```

```python
results=results
cap = cv2.VideoCapture("xyz.mp4")

count = 0
line_y = 254  
tracker_memory = {}  # {id: last_cy}
counted_ids = set()   # IDs that already crossed

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    if results[0].boxes.id is not None:
            # Get boxes, IDs, and classes
            boxes = results[0].boxes.xyxy.cpu().numpy()
            ids = results[0].boxes.id.cpu().numpy().astype(int)  # Converting tensors to array and numpy runs on cpu
            
            for box, obj_id in zip(boxes, ids):
                x1, y1, x2, y2 = box
                cx, cy = int((x1 + x2) / 2), int((y1 + y2) / 2)

                # 3. Crossing Logic
                if obj_id in tracker_memory:
                    prev_cy = tracker_memory[obj_id]

                    # If center point moves from above line to below line
                    if prev_cy < line_y and cy >= line_y:
                        if obj_id not in counted_ids:
                            count += 1
                            counted_ids.add(obj_id)

                # Update memory
                tracker_memory[obj_id] = cy

                # Visuals
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
                cv2.putText(frame, f"ID: {obj_id}", (int(x1), int(y1) - 10), 
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
                cv2.circle(frame, (cx, cy), 4, (0, 0, 255), -1)

        # Draw the counting line and total count
        cv2.line(frame, (0, line_y), (frame.shape[1], line_y), (0, 255, 255), 2)
        cv2.putText(frame, f"Total Vehicles: {count}", (30, 50), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)

    cv2.imshow("YOLO Tracking & Counting", frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
